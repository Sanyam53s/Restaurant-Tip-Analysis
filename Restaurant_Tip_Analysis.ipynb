### Data Analysis Project: Restaurent Tips Analysis 


## objectives
- Do male tip more than female?
- Does Day affect tip?
- Does bigger group give larger tips ?
- Does time (Lunch/Dinner) affective? 

```python
# Dataset
import seaborn as sns
df = sns.load_dataset('tips')
df.head(3)
```

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Gather unformation about Dataset 
df.info()
```

```python
#Dataset's Numerical values Varience 
df.describe()
```

```python
#Total null values in a dataset
print('Null Values',df.isnull().sum().sum())
```

```python
# Columns in Dataset
df.columns
```

##### Total_Bill Distribution

```python
plt.figure( figsize= (6,3))
sns.histplot(df['total_bill'], kde=True)
plt.title("Total Bill Distribution ")
plt.show()
```

##### Total_Bill vs Tips

```python
plt.figure(figsize = (6,3))
sns.scatterplot( x='total_bill', y='tip', data=df)
plt.title("Total_Bill VS Tips")
plt.xlabel("Total_bill")
plt.ylabel("Tips")
plt.show()
```

##### Tips by Gender

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='sex',y='tip',data=df)
plt.title('Tips by gender')
plt.xlabel('Gender')
plt.ylabel('Tip')
plt.show()
```

##### Tips By Day


```python
plt.figure(figsize=(6,3))
sns.boxplot(x='day',y='tip',data=df, palette ='coolwarm' )
plt.title('Tips by Day')
plt.xlabel('Gender')
plt.ylabel('Tip')
plt.show()
```

##### Tips by Time

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='time',y='tip',data=df, palette ='viridis' )
plt.title('Tips by time')
plt.xlabel('Time')
plt.ylabel('Tip')
plt.show()
```

##### Pairplot Analysis

```python
sns.pairplot(df)
```

##### Correlation Heatmap Analysis

```python
corr=df.corr(numeric_only = True)
plt.figure(figsize = (6,3))
sns.heatmap(corr,annot=True)
plt.show()
```

#### Insights
- Higher Total_bill -> Higher_Tips
- Dinner tip s higher than lunch
- Group size affect Total_bill and Tips
- Weekends Tips are Higher
